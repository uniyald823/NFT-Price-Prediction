```python
train_generator = ImageDataGenerator(rotation_range=2, horizontal_flip=True, zoom_range=.1)

val_generator = ImageDataGenerator(rotation_range=2, horizontal_flip=True, zoom_range=.1)

test_generator = ImageDataGenerator(rotation_range=2, horizontal_flip= True, zoom_range=.1)
```

```python
train_generator.fit(X_train)
val_generator.fit(X_val)
test_generator.fit(X_test)
```

```python
model3= Sequential()
model3.add(base_model) 
model3.add(Flatten())
model3.add(Dense(1024,activation=('relu'),input_dim=512))
model3.add(Dense(512,activation=('relu'))) 
model3.add(Dense(256,activation=('relu'))) 
model3.add(Dropout(.3))
model3.add(Dense(128,activation=('relu')))
#model.add(Dropout(.2))
model3.add(Dense(7,activation=('softmax'))) 

#Checking the final model summary
model3.summary()
```

```python
batch_size= 10
epochs=10
learn_rate=.001
sgd=SGD(learning_rate=learn_rate,momentum=.9,nesterov=False)
adam=Adam(learning_rate=learn_rate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model3.compile(optimizer=sgd,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model3.fit_generator(train_generator.flow(X_train, Y_train, batch_size= batch_size),epochs = epochs, steps_per_epoch = X_train.shape[0]//batch_size, validation_data = val_generator.flow(X_val, Y_val, batch_size = batch_size), validation_steps = 10, callbacks=[lrr], verbose = 1)
```

```python
model2 = Sequential()

model2.add(Conv2D(filters = 16, kernel_size = (3, 3), activation='relu',
                 input_shape = (64,64,3)))
model2.add(BatchNormalization())
model2.add(Conv2D(filters = 16, kernel_size = (3, 3), activation='relu'))
model2.add(BatchNormalization())
model2.add(MaxPool2D(strides=(2,2)))
model2.add(Dropout(0.1))
model2.add(Conv2D(filters = 32, kernel_size = (3, 3), activation='relu'))
model2.add(BatchNormalization())
model2.add(Conv2D(filters = 32, kernel_size = (3, 3), activation='relu'))
model2.add(BatchNormalization())
model2.add(MaxPool2D(strides=(2,2)))
model2.add(Dropout(0.1))

model2.add(Flatten())
model2.add(Dense(512, activation='relu'))
model2.add(Dropout(0.1))

model2.add(Dense(1024, activation='relu'))
model2.add(Dropout(0.1))
model2.add(Dense(n_classes, activation='softmax'))

learning_rate = 0.001

model2.compile(loss = categorical_crossentropy,
              optimizer = Adam(learning_rate),
              metrics=['accuracy'],run_eagerly=True)

model2.summary()
```

```python
save_at = "model2.hdf5"
save_best = ModelCheckpoint (save_at, monitor='val_accuracy', verbose=0, save_best_only=True, save_weights_only=False, mode='max')
```

```python
history = model2.fit( X_train, Y_train, 
                    epochs = 10, batch_size = 100, 
                    callbacks=[save_best], verbose=1, 
                    validation_data = (X_val, Y_val))

# Plot the training history (Training accuracy & Validation accuracy)

plt.figure(figsize=(6, 5))
plt.plot(history.history['accuracy'], color='r')
plt.plot(history.history['val_accuracy'], color='b')
plt.title('Model Accuracy', weight='bold', fontsize=16)
plt.ylabel('accuracy', weight='bold', fontsize=14)
plt.xlabel('epoch', weight='bold', fontsize=14)
plt.ylim(0.4, 0.9)
plt.xticks(weight='bold', fontsize=12)
plt.yticks(weight='bold', fontsize=12)
plt.legend(['train', 'val'], loc='upper left', prop={'size': 14})
plt.grid(color = 'y', linewidth='0.5')
plt.show()
```
